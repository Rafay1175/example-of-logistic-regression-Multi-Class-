```python
# importing the libraries
%matplotlib inline 
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
digits = load_digits() # assigning the load_digits function to digit variable
```

```python
dir(digits) # dir functions gets all the methods of a variable
```

```python
digits.data[0] # returns an array of size 64 for a digit
```

```python
plt.gray() # this will set the the color of the graph to gray 
plt.matshow(digits.images[0]) # matshot function is used to display the visual image of an array 
```

```python
digits.target[0:5] # target method is used to check the value of a given number
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train , X_test  , y_train , y_test = train_test_split(digits.data , digits.target , test_size = 0.2 , random_state = 42)
```

```python
len(X_train)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter= 10000) # set the limit of iterator to 10000 so that it does not run out of space
```

```python
model.fit(X_train , y_train) #training the model
```

```python
model.score(X_test , y_test) # to check how accurate our model is
```

```python
digits.target[180] # to check the actual value to compare it to the predicted value of our model
```

```python
model.predict(digits.data[0:5]) # some predicted values 
```

```python
y_predicted = model.predict(X_test) # making a new variable for our model prediction values 
from sklearn.metrics import confusion_matrix # importing the cofusion_matrix module from sklearn.metrics

matrix = confusion_matrix(y_test , y_predicted) # this will create a 2d array matrix for actual vs predicted values

matrix
```

```python
len(matrix)
```

```python
import seaborn as sns
plt.figure(figsize = (10,7) , dpi = 500) # figsize = (length x width) of the plot and dpi is dots per inch
sns.heatmap(matrix , annot = True , cmap = "YlGnBu") # this will create a heatmap for out matrix
plt.xlabel("predicted") 
plt.ylabel("actual") 
```
